# pytsmp: A Python library for matrix profile

## Load libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from pytsmp import STAMP, STOMP
```

## Create a random walk

```python
ts = np.cumsum(np.random.randint(2, size=(5000,)) * 2 - 1)
```

```python
plt.plot(ts)
```

## Specify a window size and run STAMP algorithm

```python
w = 500
mp_stamp = STAMP(ts, window_size=w, verbose=True)
```

```python
# get the profiles
mpro_stamp, ipro_stamp = mp_stamp.get_profiles()
```

## Plot the matrix profile

```python
plt.plot(ts, label="Data")
plt.plot(mpro_stamp, label="STAMP profile")
plt.legend(loc="upper right")
plt.show()
```
